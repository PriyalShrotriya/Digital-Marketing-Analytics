```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as BS
import pandas as pd
from urllib.parse import urlparse, urlsplit
```

```python
website_page = 'http://www.bizrate.com/mattresses-boxsprings/hybrid-10-inches-medium-firm-mattress/full/663963-270259/buy?maxPrice=430&minPrice=410'
page = urlopen(website_page)
soup = BS(page)
```

```python
div_finds = soup.find_all('div',class_='catalog results')
```

```python
title_list=[]
cost_list=[]
link_list=[]
for div in div_finds:
    tables = div.find_all('td',class_='details details-middle')
    for table in tables:
        para = table.find_all('p',class_='title')
        for p in para:
            title_list.append(p.text)
    costandsite = div.find_all('td',class_='purchase purchase-center')
    for cost in costandsite:
        costp = cost.find_all('a')[0]
        cost_list.append(costp.text)
    shoplinks = div.find_all('div',class_='merchant-details-with-price')
    for shop in shoplinks:
        link = shop.find_all('a')[0]
        for l in link:
            link_list.append(l)
```

```python
final_merch_list=[]
for ll in link_list:
    if len(ll) > 0:
        final_merch_list.append(ll)
    else:
        final_merch_list.append(ll['title'])
```

```python
import pandas as pd
finaldf=pd.DataFrame()
finaldf['ProductName']=title_list
finaldf['Price']=cost_list
finaldf['Seller']=final_merch_list
finaldf
```

```python
finaldf['Price'] = finaldf.Price.apply(lambda x: x.strip('$'))
```

```python
#price = 420
def compare(seller_price_list, price): 
      
    # traverse in the list 
    for x in seller_price_list: 
        
        # compare with all the values 
        # with val 
        if price >= x: 
            return False 
    return True
```
